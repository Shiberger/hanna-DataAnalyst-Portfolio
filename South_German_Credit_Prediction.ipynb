<h1>จัดการกับ Outlier</h1>

```python
def detect_outliers(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
```

```python
# หาค่า outlier ใน 'amount', 'duration', 'age'
outliers_amount = detect_outliers(df_train, 'amount')
outliers_duration = detect_outliers(df_train, 'duration')
outliers_age = detect_outliers(df_train, 'age')
```

```python
print(f"Number of outliers in 'amount': {len(outliers_amount)}")
print(f"Number of outliers in 'duration': {len(outliers_duration)}")
print(f"Number of outliers in 'age': {len(outliers_age)}")
```

```python
import numpy as np
from sklearn.preprocessing import RobustScaler
```

```python
# ตรวจสอบค่า Skewness ของฟีเจอร์ 'amount', 'duration', 'age'
skew_amount = df_train['amount'].skew()
skew_duration = df_train['duration'].skew()
skew_age = df_train['age'].skew()

print(f"Skewness of 'amount': {skew_amount}")
print(f"Skewness of 'duration': {skew_duration}")
print(f"Skewness of 'age': {skew_age}")
```

```python
#Log Transform สำหรับฟีเจอร์ที่ skewed
df_train['log_amount'] = np.log1p(df_train['amount'])
df_train['log_duration'] = np.log1p(df_train['duration'])
df_train['log_age'] = np.log1p(df_train['age'])
```

```python
#ใช้ RobustScaler สำหรับข้อมูลที่มีค่า outlier แรงๆ เช่น 'amount', 'duration'
scaler = RobustScaler()
df_train['scaled_amount'] = scaler.fit_transform(df_train[['amount']])
df_train['scaled_duration'] = scaler.fit_transform(df_train[['duration']])
df_train['scaled_age'] = scaler.fit_transform(df_train[['age']])
```

```python
#สร้างฟีเจอร์ is_outlier
df_train['is_outlier_amount'] = (df_train['amount'] > df_train['amount'].quantile(0.75) + 1.5 * (df_train['amount'].quantile(0.75) - df_train['amount'].quantile(0.25))) | (df_train['amount'] < df_train['amount'].quantile(0.25) - 1.5 * (df_train['amount'].quantile(0.75) - df_train['amount'].quantile(0.25)))

df_train['is_outlier_duration'] = (df_train['duration'] > df_train['duration'].quantile(0.75) + 1.5 * (df_train['duration'].quantile(0.75) - df_train['duration'].quantile(0.25))) | (df_train['duration'] < df_train['duration'].quantile(0.25) - 1.5 * (df_train['duration'].quantile(0.75) - df_train['duration'].quantile(0.25)))

df_train['is_outlier_age'] = (df_train['age'] > df_train['age'].quantile(0.75) + 1.5 * (df_train['age'].quantile(0.75) - df_train['age'].quantile(0.25))) | (df_train['age'] < df_train['age'].quantile(0.25) - 1.5 * (df_train['age'].quantile(0.75) - df_train['age'].quantile(0.25)))
```

<h1>Train model

```python
from sklearn.model_selection import train_test_split

X = df_selected_clean.drop(columns=['credit_risk', 'is_outlier_amount', 'is_outlier_duration', 'is_outlier_age'])
y = df_selected_clean['credit_risk']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17, stratify=y)
```

```python
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

model = LogisticRegression(random_state=17)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

dt_model = DecisionTreeClassifier(random_state=17)

dt_model.fit(X_train, y_train)

y_pred_dt = dt_model.predict(X_test)

accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Decision Tree Accuracy: {accuracy_dt:.4f}")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
print("\nClassification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
```

```python
smote = SMOTE(random_state=17)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

log_reg = LogisticRegression(random_state=17)
grid_search = GridSearchCV(log_reg, param_grid, cv=5, n_jobs=-1, scoring='accuracy', verbose=1)
grid_search.fit(X_train_resampled, y_train_resampled)

print("Best parameters for Logistic Regression:", grid_search.best_params_)

y_pred_log_reg = grid_search.best_estimator_.predict(X_test)

accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
print(f"Accuracy: {accuracy_log_reg:.4f}")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_log_reg))
print("\nClassification Report:\n", classification_report(y_test, y_pred_log_reg))
```

```python
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=17)
X_train_under, y_train_under = undersampler.fit_resample(X_train, y_train)

log_reg.fit(X_train_under, y_train_under)

y_pred_under = log_reg.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_under))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_under))
print("Classification Report:\n", classification_report(y_test, y_pred_under))
```
